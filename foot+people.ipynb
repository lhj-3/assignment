```python
import cv2
import mediapipe as mp
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
mp_objectron = mp.solutions.objectron
from google.colab.patches import cv2_imshow
```

```python
count = 0
count_pe = 0
cap = cv2.VideoCapture('/content/human.mp4')
cap_foot = cv2.VideoCapture('/content/shoe_off.mp4')
with mp_pose.Pose(min_detection_confidence=0.8,min_tracking_confidence=0.8) as pose,mp_objectron.Objectron(static_image_mode=False,max_num_objects=1,min_detection_confidence=0.5,min_tracking_confidence=0.99,model_name='Shoe') as objectron :
  while (cap.isOpened() and cap_foot.isOpened()):
    success, image = cap.read()
    success_foot, image_foot = cap_foot.read()
    if (not success) or (not success_foot):
      print("Ignoring empty camera frame.")
      # If loading a video, use 'break' instead of 'continue'.
      break

    image = image[0 : 380, 100 : 752]

    # Flip the image horizontally for a later selfie-view display, and convert
    # the BGR image to RGB.
    image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
    image_foot = cv2.cvtColor(image_foot, cv2.COLOR_BGR2RGB)
    
    # To improve performance, optionally mark the image as not writeable to
    # pass by reference.
    image.flags.writeable = False
    image_foot.flags.writeable = False
    results = pose.process(image)
    results_foot = objectron.process(image_foot)

    # Draw the pose annotation on the image.
    image.flags.writeable = True
    image_foot.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    image_foot = cv2.cvtColor(image_foot, cv2.COLOR_RGB2BGR)

    if results_foot.detected_objects:
      for detected_object in results_foot.detected_objects:
        mp_drawing.draw_landmarks(image_foot, detected_object.landmarks_2d, mp_objectron.BOX_CONNECTIONS)
        mp_drawing.draw_axis(image_foot, detected_object.rotation,detected_object.translation)
#    if (results.pose_landmarks != None):
#      print("pose")

    x2 = (detected_object.landmarks_2d.landmark[2].x)
    y2 = (detected_object.landmarks_2d.landmark[2].y)
    x1 = (detected_object.landmarks_2d.landmark[1].x)
    y1 = (detected_object.landmarks_2d.landmark[1].y)
    slope = (y2-y1)/(x2-x1)

    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
    cv2.imwrite("/content/image/frame%d.jpg" % count, image)
    cv2.imwrite("/content/result/frame%d.jpg" % count, image_foot)
    count += 1
    if((results.pose_landmarks != None) and (abs(slope) > 0.3)):
      print("stop")

cap.release()
cap_foot.release()
```
